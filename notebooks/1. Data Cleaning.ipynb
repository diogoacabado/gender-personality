### Column: gender

```python
gender = {0: 'Unknown',
          1: 'Male', 
          2: 'Female',
          3: 'Other'}
```

```python
data['gender'].replace(gender, inplace=True)
```

```python
filt = data['gender'] == 'Unknown'
n = data.loc[filt, 'gender'].count()
m = data.shape[0]
p = 100*n/m
percentage = round(p,2)

print('percentage:', str(percentage) + '%')
```

```python
filt = data['gender'] == 'Other'
n = data.loc[filt, 'gender'].count()
m = data.shape[0]
p = 100*n/m
percentage = round(p,2)

print('percentage:', str(percentage) + '%')
```
