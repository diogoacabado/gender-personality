```python
table = pd.read_csv('../output_data/questions_table.csv')
table.head()
```

```python
data = pd.read_csv('../output_data/data_cleaned.csv')
data.head()
```

```python
filt = list(data.columns)[1:]
aux = data[filt]
```

```python
aux = aux*list(table['key'])
aux.head()
```

```python
filt = list(aux.columns)[30:40]
aux[filt]
```

```python
traits = ['Extraversion', 'Neuroticism', 'Agreeableness', 'Conscientiousness', 'Openness']
print(traits)
```

```python
filt = list(aux.columns)
seq = lambda x: filt[10*x:10*x + 10]
```
